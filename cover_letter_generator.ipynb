```python
import os
import requests

def generate_cover_letter(cv, fiche_poste, api_key=None, model="mistral-medium", temperature=0.3):
    # Retrieve the API key from the environment if not provided
    if not api_key:
        api_key = os.getenv("MISTRAL_API_KEY")
    if not api_key:
        raise ValueError("API key not found. Please set the MISTRAL_API_KEY environment variable or pass it as an argument.")

    # Mistral API endpoint
    url = "https://api.mistral.ai/v1/chat/completions"

    # Create the prompt
    prompt = f"""
    Generate a professional cover letter for the following job description:

    {fiche_poste}

    Use the following CV details to tailor the cover letter:

    {cv}

    The cover letter should be professional and directly highlight the relevant skills and experiences.
    """

    # Data payload for the API request
    data = {
        "model": model,
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "temperature": temperature
    }

    # Send the request
    response = requests.post(
        url,
        headers={"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"},
        json=data
    )

    # Return the generated text or handle errors
    if response.status_code == 200:
        result = response.json()
        return result["choices"][0]["message"]["content"]
    else:
        raise Exception(f"Error: {response.status_code}, {response.text}")

# Example usage
cv = """
Alexis Vannson
French and Austrian nationality | vannson.alexis@gmail.com | www.linkedin.com/in/alexis-vannson

EDUCATION
- CentraleSup√©lec - Essec Business School: B.S. in Artificial Intelligence, Data and Management Sciences, 2nd year student.
  Relevant coursework: Programming for Data Science, Algorithms, Data Structures, Data Analysis, Optimization, Applied Linear Algebra, Calculus.

- 42 Paris: Focused on C, Shell scripting, Python, UNIX/Linux, command line, Git, GitHub, Makefile, memory management, Algorithms & Data Structures.

EXPERIENCE
- **Emeria Technologies** (April 2024): Implemented a predictive Random Forest model using Python and Scikit-learn, achieving 85% accuracy.
- **Machine Learning Research Assistant, CentraleSup√©lec** (March-June 2024): Comparative analysis of MLPs, CNNs, GNNs, and UNet architectures using PyTorch.

PERSONAL PROJECTS
- **Pneumonia Classification**: Developed a CNN in PyTorch achieving 86% accuracy. Implemented a Random Forest classifier with Scikit-learn achieving 96% accuracy.

SKILLS
- Programming: Python, SQL, C, HTML/CSS, JavaScript.
- Data Science: NumPy, Pandas, Scikit-learn, PyTorch, TensorFlow, Matplotlib.
- Languages: French (Native), German (Native), English (Fluent), Spanish (Basic), Dutch (Basic).

INTERESTS
- Hockey: Represented France in the U18 National Field Hockey Team (2021-2022).
"""

fiche_poste = """
√Ä propos de l‚Äôoffre d‚Äôemploi
ML Engineer - GenAI (Intern)

üìç Location: Bordeaux, France

üïí Type: Intern

üõ†Ô∏è Tech Stack: Python

About Klark

Founded in May 2023, Klark‚Äôs mission is to revolutionize customer service through generative AI.

As pioneers in the AI industry, we‚Äôre developing a game-changing software solution for customer care contact management. Our cutting-edge technology offers a unique opportunity to have a significant, direct impact on how companies interact with their customers.

Join us and contribute significantly to an industry poised for transformation!

Job Overview

We are in search of a ML Engineer to help take our generative AI capabilities to the next level.

Under the supervision of our ML Engineer, you will work in an agile, AI-driven environment where your contributions will have a direct and crucial impact on our business.

Your role will be essential in developing AI algorithms that will serve as a pillar for our Generative Customer Service solution.

Responsibilities

- R&D: Conduct in-depth research on state-of-the-art topics such as NLP, retrievers, LLMs, and Prompt Engineering with a significant impact on our startup.
- Modeling: Develop, train, and monitor retrievers and LLMs models tailored to our business data.
- Advanced RAG: Implement new RAG (Retrieval Augmented Generation) techniques and orchestration of LLM agents.
- NLP use cases: Implement ML algorithms to NLP tasks such as text classification, clustering, NER (Named Entity Recognition), etc.
- Collaboration: Work closely with other developers for seamless model deployment.

Qualifications

- Experience: Graduate from top engineering schools (X, Centrale, Ponts, Mines, Telecom, etc.) Preferably in final year of studies.
- Technical Skills: Expertise in Python, NLP, language models, machine learning algorithms, libraries like PyTorch, transformers, etc. Knowledge of frameworks such as LangChain, DSPy, vLLM, etc. is a plus.
- Soft Skills: Excellent communication, synthesis abilities, problem-solving skills, and teamwork capabilities.
"""

# Generate and print the cover letter
try:
    cover_letter = generate_cover_letter(cv, fiche_poste)
    print(cover_letter)
except Exception as e:
    print(e)
```
